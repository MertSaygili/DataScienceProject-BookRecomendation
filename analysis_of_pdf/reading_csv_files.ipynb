```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from sklearn.preprocessing import LabelEncoder
from wordcloud import WordCloud, STOPWORDS  # pip install wordcloud
```

```python
# gets datas inside of csv files
main_path = 'pdf_csv_files/'

world_war_df = pd.read_csv(f'{main_path}WorldWar1.csv')
medical_df = pd.read_csv(f'{main_path}Medical.csv')
music_df = pd.read_csv(f'{main_path}Music.csv')
man_woman_relationship_df = pd.read_csv(f'{main_path}Man-woman relationships.csv')
education_df = pd.read_csv(f'{main_path}Education.csv')
cooking_df = pd.read_csv(f'{main_path}Cooking.csv')
astronomy_df = pd.read_csv(f'{main_path}Astronomy.csv')
religion_df = pd.read_csv(f'{main_path}Religion.csv')
science_df = pd.read_csv(f'{main_path}Science.csv')
```

```python
def word_analysis(category):
    stopwords = list(STOPWORDS)

    title_w = ''

    for word in category.word:
        word = word.lower()
        splited = re.findall(r'\b[A-Za-z]+\b', word)
        splited = [w for w in splited if w not in stopwords]
        title_w += ' '.join(splited)+ ' '


    wordcloud = WordCloud(width = 1100, 
                        height = 800, 
                        background_color ='white', 
                        stopwords = stopwords, 
                        min_font_size = 8).generate(title_w) 
    
    # plot the WordCloud image                        
    plt.figure(figsize = (6, 8), 
            facecolor = None) 

    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 
    
    plt.show() 


def word_analysis_pie(category):
    plt.pie(category['count'], labels=category['word'])
    plt.set_cmap('Spectral')
    plt.figure(figsize=(20, 20))
    plt.show()
    
def word_analysis_barh(category):
    plt.barh(category['word'], category['count'])
    plt.title('Word relationships with average word reoeat')
```

```python
word_analysis(world_war_df)
```

```python
word_analysis_pie(world_war_df)
```
