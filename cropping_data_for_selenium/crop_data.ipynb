```python
import pandas as pd
import numpy as np
import csv
```

```python
data = pd.read_csv('../inputs/Preprocessed_data.csv') # reads data
data = data.drop_duplicates(subset=['book_title']) # drops dublicate rows
data = data.query("Language == 'en'") # picking english books only

# drop unuse columns
data = data.drop(['user_id', 'location', 'age', 'isbn', 'rating', 'year_of_publication', 'publisher', 'img_s', 'img_m', 'img_l', 'Summary', 'city', 'state', 'country', 'Language'], axis=1)
data = data.iloc[:, 1:]
data.head(20) # showing first 20 rows
```

```python
all_categories = data['Category'] # gets Categories only

unique_categories = [] # will save unique categories

for category in all_categories:
    if not unique_categories.__contains__(category): # if category is unique then add that category yo unique_cateogry list  
        unique_categories.append(category)


# printing unique categories
for category in unique_categories:
    print(category)
```

```python
# for every category and get maximum 50 book
limit = 50 # defining limit
column_count = 3
unique_categories_info = []


for category in unique_categories:
    temp_data = data[data['Category'] == category]
    if temp_data.size / column_count > limit:
        unique_categories_info.append(temp_data.head(limit))
    else:
        unique_categories_info.append(temp_data)
```

```python
# creates csv file

fields = ['book_title', 'book_author', 'Category'] # fields
rows = [] # rows
rows.append(fields) # adds feature names to rows list

for category in unique_categories_info:
    for i in range(0, len(category)):
        # add every row into rows list
        tempRow = [category['book_title'].iloc[i], category['book_author'].iloc[i], category['Category'].iloc[i]]
        rows.append(tempRow)


filename = '../outputs/cropped_data/cropped_data.csv' # defines file name

with open(filename, 'w', encoding='utf-8', newline='') as file: # opens file and writes rows list into it
    writer = csv.writer(file)
    writer.writerows(rows)

```
