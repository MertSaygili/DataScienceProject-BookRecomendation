```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
predict_rate_df = pd.read_csv('predict_rate.csv') 

df = predict_rate_df.query('rates != 0.000000')
df

# The table says that, this book belongs 15 percent architectural books 
# The table says that, this book belongs 15 percent art books 
# The table says that, this book belongs 15 percent Astronomy books 
# The table says that, this book belongs 15 percent Biography & Autobiography books 
# The table says that, this book belongs 15 percent Business & Ecenomics books 
# The table says that, this book belongs 15 percent Childrens books  and so on
```

```python
plt.pie(df['rates'], labels=df['category_names'], startangle=0, autopct='%.1f')
plt.show()
```

```python
plt.barh(df['category_names'], df['rates'])
plt.title('Category relationships with rate')
plt.show()
```
